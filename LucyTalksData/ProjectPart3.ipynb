```python
%%bigquery --project data-viz-337501
SELECT 
  state_name,
  SUM(number_of_fatalities) as total_deaths
FROM `bigquery-public-data.nhtsa_traffic_fatalities.accident_2016`
WHERE state_name is not null
GROUP BY state_name
```

Now I am going to use this table to compare to the averate value

```python
%%bigquery --project data-viz-337501
WITH fatalities as (
SELECT 
  state_name,
  SUM(number_of_fatalities) as total_deaths
FROM `bigquery-public-data.nhtsa_traffic_fatalities.accident_2016`
WHERE state_name is not null
GROUP BY state_name)

SELECT *
FROM fatalities
WHERE total_deaths < (
  SELECT 
    AVG(total_deaths)
  FROM 
    fatalities)
ORDER BY state_name
```
